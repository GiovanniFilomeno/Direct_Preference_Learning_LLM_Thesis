```python
import matplotlib.pyplot as plt
import os, sys

current_dir = os.getcwd()
module_path = os.path.abspath(os.path.join(current_dir, "../src"))

if module_path not in sys.path:
    sys.path.append(module_path)

from maze import Maze
from maze_env import draw_map
import numpy as np

maze_size = 10
start = np.array([0.05, 0.05])     # punto di partenza (blu)
goal  = np.array([0.95, 0.85])     # punto di arrivo (rosso)

maze = Maze(maze_size, maze_size, 0, 0)
maze.make_maze_fail(step=3) 

fig, ax = plt.subplots(figsize=(6, 6))

# Disegna la mappa
draw_map(1 / maze_size, maze, ax=ax, alpha=1.0)

# Disegna la partenza (cerchio blu)
ax.plot(start[0], start[1], 'bo', markersize=10, label='Start')

# Disegna l'arrivo (croce rossa)
ax.plot(goal[0], goal[1], 'rx', markersize=10, label='Goal')

# Setup grafico
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.set_aspect('equal')
ax.invert_xaxis()
ax.legend(loc="upper right")
plt.title("Maze with Start (blu) e Goal (red)", fontsize=21)
plt.show()
```
